1. Key Features of Python
- Python is a popular programming language due to the following features:

- Easy to Learn and Use: Python has a simple syntax resembling English, making it beginner-friendly.

- Interpreted Language: Python code is executed line-by-line, which makes debugging easier.

- Dynamically Typed: You don’t need to declare variable types explicitly.

- Extensive Standard Library: Offers built-in modules and functions to perform many tasks.

- Object-Oriented: Supports classes and objects for modular, reusable code.

- Platform Independent: Python code can run on different operating systems without modification.

- Large Community Support: Rich ecosystem of libraries and a helpful community.

2. Predefined Keywords in Python
- Keywords are reserved words in Python that have special meanings. They cannot be used as identifiers (variable names).

- Examples of Keywords: if, else, for, while, def, class, return, True, False, None, in, is

- Example Program:
    def greet(name):
    if name:
        return "Hello, " + name
    else:
        return "Hello, Stranger"

  print(greet("Alice"))
  print(greet(""))
  
  Explanation:

  def defines a function.

  if/else are used for conditional branching.
  return sends back a result from the function.

3. Mutable vs Immutable Objects
- Feature	Mutable	Immutable
Can be changed	Yes	No
Examples	List, Dict, Set	int, float, str, tuple
Memory location	Same object (ID)	New object (ID changes)

Mutable Example:

my_list = [1, 2, 3]
my_list.append(4)
print(my_list)  # [1, 2, 3, 4]

 4. Type Casting in Python
 - Type casting is converting one data type into another.

 Types:

 int(): Converts to integer

 float(): Converts to float

 str(): Converts to string

 list(), tuple(), set() for collection conversions

 Examples:


 x = "123"
 y = int(x)       # '123' → 123
 z = float(y)     # 123 → 123.0
 print(type(y), type(z))
 # <class 'int'> <class 'float'>

 a = 5
 b = str(a)       # 5 → "5"
 print(type(b))   # <class 'str'>

 5. Types of Operators in Python
 - Python provides various operators:

 Type	Operators	Example
 Arithmetic	+, -, *, /, %, **	5 + 2 = 7
 Comparison	==, !=, <, >, <=, >=	5 > 3 = True
 Logical	and, or, not	True and False = False
 Assignment	=, +=, -=, etc.	x += 1
 Bitwise	&, `	, ^, ~, <<, >>`
 Membership	in, not in	'a' in 'apple'
 Identity	is, is not	x is y

 Example:
 a = 10
 b = 5
 print(a + b)       # 15 (Arithmetic)
 print(a > b)       # True (Comparison)
 print(a > 5 and b < 10)  # True (Logical)

 6. Conditional Statements in Python
- Used to make decisions in code.

 Syntax:

 if condition:
    # code block
 elif another_condition:
    # code block
 else:
    # code block
 Example:

 age = 18
 if age < 18:
    print("Minor")
 elif age == 18:
    print("Just became an adult")
 else:
    print("Adult")

 7. Loops in Python
- For Loop
 Iterates over a sequence (like list, string, range).

 Example:

 for i in range(5):
    print(i)  # 0 to 4
 While Loop
 Repeats while a condition is True.

Example:

 count = 0
 while count < 5:
    print(count)
    count += 1
